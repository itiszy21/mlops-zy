## Task 1: Exploratory Data Analysis

```python
# import the pandas library
import pandas as pd
```

```python
#load in the dataset and read it
resort_df = pd.read_csv("01_homely_resort_listing.csv")
resort_df.head()
```

```python
#checking the data type (before data cleaning)
resort_df.info()
```

```python
#remove the "Unnamed: 25" column
resort_df = resort_df.drop(columns=["Unnamed: 25"])
resort_df.head()
```

```python
import numpy as np

# Identify columns with NaN values
nan_columns = ['review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_rating', 'review_scores_value']

# Create a boolean mask for rows with NaN values in specified columns
nan_mask = resort_df[nan_columns].apply(lambda col: np.isnan(col), axis=1).any(axis=1)

# Filter out rows with NaN values
resort_df = resort_df[~nan_mask]

# Display the first few rows of the cleaned DataFrame
resort_df.head()
```

```python
#converting the "price" column into numeric data type, removing the "$" and ","

resort_df['price'] = resort_df['price'].replace('[\$,]', '', regex=True).astype(float)

resort_df_cleaned = resort_df
resort_df_cleaned.head()
```

```python
#check data type (after data cleaning)
resort_df_cleaned.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#checking the correlation for the different variables

plt.figure(figsize=(12, 8))
correlation_matrix = resort_df_cleaned.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# List of columns to remove due to low correlation
columns_to_remove = ['availability_30', 'calculated_host_listings_count', 'guests_included',
                      'host_listings_count', 'latitude(North)', 'longitude(East)', 'maximum_nights',
                      'number_of_reviews', 'review_scores_checkin', 'review_scores_communication',
                      'review_scores_location', 'review_scores_rating', 'review_scores_value']

# Drop the specified columns
resort_df_cleaned_2 = resort_df_cleaned.drop(columns=columns_to_remove)

# Display the first few rows of the updated DataFrame
resort_df_cleaned_2.head()
```

```python
#correlation after removing certain variables with low correlation

plt.figure(figsize=(12, 8))
correlation_matrix = resort_df_cleaned_2.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#histogram of the distribution of prices

plt.figure(figsize=(10, 6))
sns.histplot(resort_df_cleaned_2['price'], bins=30, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#boxplot of the price by room type

plt.figure(figsize=(12, 8))
sns.boxplot(x='room_type', y='price', data=resort_df_cleaned_2)
plt.title('Boxplot of Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

```python
#countplot of the property types 

plt.figure(figsize=(12, 8))
sns.countplot(x='property_type', data=resort_df_cleaned_2, order=resort_df_cleaned_2['property_type'].value_counts().index)
plt.title('Countplot of Property Types')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Task 2: Train, validate and develop a Machine Learning pipeline using PyCaret

```python
# import pycaret regression and init setup

from pycaret.regression import *
import mlflow

s = setup(resort_df_cleaned_2, target = "price", session_id=123, fold=10, normalize=True,
          fold_strategy='kfold', verbose=False, train_size=0.8,
          transformation = True, transform_target = True,
          #remove_multicollinearity = True, multicollinearity_threshold = 0.95,
          log_experiment = True, log_plots = True, experiment_name = 'Resort')
          #bin_numeric_features=["accommodates", "bathrooms", "bedrooms", "beds"])

mlflow.set_tracking_uri("http://127.0.0.1:5000")
tracking_uri = mlflow.get_tracking_uri()
print("current tracking uri: {}".format(tracking_uri))

#compare baseline models
best = compare_models()
```

```python
# plot residuals
plot_model(best, plot = 'residuals')
```

```python
# plot error
plot_model(best, plot = 'error')
```

```python
#evaluate the model

evaluate_model(best)
```

```python
#create model "ada", the model with the best r2 value

ada_model = create_model('ada', fold=10)
```

```python
#tune the ada model

tuned_ada_model = tune_model(ada_model)
```

```python
#evaluate the results of the tuned ada model

plot_model(tuned_ada_model, plot='residuals')
```

```python
# plot error of the tuned ada model
plot_model(tuned_ada_model, plot = 'error')
```

```python
#evaluate model of the tuned ada model

evaluate_model(tuned_ada_model)
```

```python
#finalize the model

final_best = finalize_model(tuned_ada_model)
```

```python
#predict on test set

holdout_pred = predict_model(tuned_ada_model)
```

```python
#shows predictions

holdout_pred.head()
```

```python
# predict model on new_resort_data
predictions = predict_model(final_best, data = new_resort_data)
predictions.head()
```
